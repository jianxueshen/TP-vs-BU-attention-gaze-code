```python
import os
import glob
import shutil
import fnmatch
import pandas as pd
import numpy as np
import cv2
import warnings
warnings.simplefilter("ignore")
```

```python
root_folder = "../RAW_DATA/RAW_DATA_OrigRes_DomEye"
videos = os.listdir(root_folder)
sz = [1920,1080]
for video in videos:
    parts = video.split("_")
    index = parts.index("RAW")
    video_name = "_".join(parts[:index])
    types = os.listdir(os.path.join(root_folder,video))
    for type in types:
        parts = type.split("_")
        tp = parts[-1]
        csvs = os.listdir(os.path.join(root_folder,video,type))
        for csv in csvs:
            parts = csv.split("_")
            index = parts.index("subj")
            sub = parts[index+1]
            
            img_name = f"fixtionmap_{video_name}_subj_{sub}_{tp}.png"
            img_folder = os.path.join("./FIXMAP",f"{video_name}_FIXMAP",f"{video_name}_FIXMAP_{tp}","Dominant")
            if not os.path.exists(img_folder):
                os.makedirs(img_folder)
            else:
                entris = os.listdir(img_folder)
                if img_name in entris:
                    continue

            data = pd.read_csv(os.path.join(root_folder,video,type,csv))
            x = data.iloc[:,-2].values
            y = data.iloc[:,-1].values
            image = np.zeros((sz[1], sz[0]), dtype=np.uint8) 
            for xi, yi in zip(x,y):
                if 0 < xi < sz[1] and 0 < yi < sz[0]:
                    image[int(xi),int(yi)] = 255
        
            img_path = os.path.join(img_folder, img_name)
            cv2.imwrite(img_path,image)
            print(f"{video_name} finised!")



```
